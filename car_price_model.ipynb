```python
# ONE HOT ENCODING 
X_encoded=pd.get_dummies(X,columns=['model','transmission','fuelType'],drop_first=True).astype(int)
X_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# 1. Reset x_label with a fresh copy of X
columns = ['model', 'transmission', 'fuelType']
x_label = X.copy()
label_encoders={}

# 2. Run the category encoder
for col in columns:
    le=LabelEncoder()

    x_label[col] = le.fit_transform(x_label[col].astype(str))
    label_encoders[col]=le

# 3. View the results
x_label
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_encoded.columns
```

```python
numeric_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
X_encoded[numeric_cols]=scaler.fit_transform(X_encoded[numeric_cols])
```

```python
x_label.columns
```

```python
x_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]=scaler.fit_transform(x_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
# Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split

# The Linear Regression model
from sklearn.linear_model import LinearRegression

# Evaluation metrics to see how well your model performed
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.33, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n=X_test.shape[0]
p=X_test.shape[1]
adjusted_r2 = 1-((1-r2)*(n-1))/(n-p-1)
adjusted_r2
```
